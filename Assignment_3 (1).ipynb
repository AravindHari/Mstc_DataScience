```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
h=pd.read_csv("C:\\Users\\HP\\Downloads\\heart.csv")
```

```python
h.head()
```

```python
sns.barplot(data=h,x="age",y="chol",hue="sex")
```

```python
sns.pairplot(h[["age","thalach","trestbps",]])
```

```python
sns.scatterplot(x="chol",y="thalach",data=h,hue="sex")
```

```python
sns.scatterplot(x="target",y="chol",data=h)
```

```python
x=h.drop("target",axis=1)
y=h["target"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier(n_neighbors=1)
```

```python
k.fit(x_train,y_train)
```

```python
pre=k.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pre))
```

```python
print(confusion_matrix(y_test,pre))
```

```python
#Implement k-Means clustering algorithm on diabetic dataset.
```

```python
d=pd.read_csv("C:\\Users\\HP\\Downloads\\diabetes.csv")
```

```python
d.head()
```

```python
sns.pairplot(d[["Age","Glucose","BMI","BloodPressure"]])
```

```python
sns.jointplot(x="Pregnancies",y="BMI",data=d,)
```

```python
sns.scatterplot(x="Glucose",y="Insulin",data=d,hue="Outcome")
```

```python
sns.jointplot(x="Glucose",y="DiabetesPedigreeFunction",data=d,kind="kde")
```

```python
from sklearn.cluster import KMeans
k=KMeans(n_clusters=2)
```

```python
k.fit(d.drop(["Outcome"],axis=1))
y_test=d["Outcome"]
```

```python
k.cluster_centers_
```

```python
k.labels_
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,k.labels_))
```

```python
print(confusion_matrix(y_test,k.labels_))
```
